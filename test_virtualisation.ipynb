```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
from matplotlib.animation import FuncAnimation
from matplotlib import animation

plt.rcParams["animation.html"] = "html5"
# Generate random data
np.random.seed(0)
X = np.random.randn(100, 2)

# Perform K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X)

# Function to update the plot for each frame of the animation
def update(frame):
    plt.cla()  # Clear the current plot

    # Plot the data points
    plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis')

    # Plot the cluster centroids
    plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', c='red', s=100)

    # Set plot title and labels
    plt.title(f'Iteration {frame+1}')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')

# Create the animation
ani = FuncAnimation(plt.gcf(), update, frames=10, interval=1000)
# Display the animation
writergif = animation.PillowWriter(fps=30)
ani.save('image/filename.gif',writer=writergif)
```
