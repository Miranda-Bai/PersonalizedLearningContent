```python
import pandas as pd
```

```python
# load data
data = pd.read_csv('data/OULAD_cleaned_data.csv')
data.shape
```

```python
# load data
vle = pd.read_csv('OULAD/studentVle.csv')
vle.shape
```

```python
joined=pd.merge(data,vle,on=['code_module','code_presentation','id_student'],how='left')
joined.shape
```

```python
joined.head()
```

```python
# adjust IMD band Oct-20 to 10-20%
joined["imd_band"] = joined["imd_band"].replace('Oct-20', '10-20%')
joined.shape
```

```python
joined[joined["imd_band"]=="10-20%"]
```

```python
joined.groupby(['code_module','code_presentation','id_student'])['sum_click'].sum()
```

```python
new_df = joined.groupby(['code_module', 'code_presentation', 'id_student'])['sum_click'].sum().reset_index()
new_df.shape
```

```python
new_df = new_df.rename(columns={'sum_click': 'total_click'})
```

```python
new_df.head()
```

```python
# adjust IMD band Oct-20 to 10-20%
data["imd_band"] = data["imd_band"].replace('Oct-20', '10-20%')
data.shape
```

```python
joinedvle=pd.merge(data,new_df,on=['code_module','code_presentation','id_student'],how='left')
```

```python
pd.DataFrame(joinedvle).to_csv("data/combineVle.csv")
```
